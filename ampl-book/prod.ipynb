### Example: prod
autogenerated using prod.mod, prod.dat, and prod.run

```python
%%writefile prod.mod
set P;

param a {j in P};
param b;
param c {j in P};
param u {j in P};

var X {j in P};

maximize Total_Profit: sum {j in P} c[j] * X[j];

subject to Time: sum {j in P} (1/a[j]) * X[j] <= b;

subject to Limit {j in P}: 0 <= X[j] <= u[j];
```

```python
%%writefile prod.dat
data;

set P := bands coils;

param:     a     c     u  :=
  bands   200   25   6000
  coils   140   30   4000 ;

param b := 40;
```

```python
%%ampl_eval
model prod.mod;
data prod.dat;
option solver cbc;
solve;
display X;
```
